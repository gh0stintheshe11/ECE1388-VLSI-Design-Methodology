```python
# Plot Internal, Switching, Total, and Leakage Power vs Frequency
import pandas as pd
import matplotlib.pyplot as plt

# Data from the updated table (units: Frequency in MHz, Power in W)
freq_mhz = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
internal_power_mw = [0.002028, 0.004056, 0.006084, 0.008112, 0.01014, 0.01217, 0.0142, 0.01622, 0.01825, 0.02028]
switching_power_mw = [0.0006917, 0.001383, 0.002075, 0.002767, 0.003459, 0.00415, 0.004842, 0.005534, 0.006225, 0.006917]
total_power_mw = [0.003512, 0.006432, 0.008952, 0.01167, 0.01439, 0.01711, 0.01983, 0.02255, 0.02527, 0.02799]
leakage_power_mw = [0.0007924, 0.0007924, 0.0007924, 0.0007924, 0.0007924, 0.0007924, 0.0007924, 0.0007924, 0.0007924, 0.0007924]

# Build DataFrame and compute total power = internal + switching + leakage
_df = pd.DataFrame({
    "Frequency (MHz)": freq_mhz,
    "Internal Power (mW)": internal_power_mw,
    "Switching Power (mW)": switching_power_mw,
    "Total Power (mW)": total_power_mw,
    "Leakage Power (mW)": leakage_power_mw
})

# Plot all series on one graph
fig, ax = plt.subplots(figsize=(9, 5.5), dpi=120)
ax.plot(_df["Frequency (MHz)"], _df["Internal Power (mW)"], marker="o", label="Internal Power")
ax.plot(_df["Frequency (MHz)"], _df["Switching Power (mW)"], marker="s", label="Switching Power")
#ax.plot(_df["Frequency (MHz)"], _df["Leakage Power (mW)"], marker="^", label="Leakage Power")
#ax.plot(_df["Frequency (MHz)"], _df["Total Power (mW)"], marker="x", label="Total Power")

ax.set_title("Power vs Frequency")
ax.set_xlabel("Frequency (MHz)")
ax.set_ylabel("Power (mW)")
ax.grid(True, which="both", linestyle="--", alpha=0.4)
ax.legend()
plt.tight_layout()
plt.show()
```
